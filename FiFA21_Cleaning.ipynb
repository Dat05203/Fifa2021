### Convert Datatime for Joined column

```python
data.Joined.value_counts()
```

```python
from numpy.core.fromnumeric import size
month_dict = {'Jan':1, 'Feb':2, 'Mar':3, 'Apr':4, 'May':5, 'Jun':6, 'Jul':7, 'Aug':8, 'Sep':9, 'Oct':10, 'Nov':11, 'Dec':12}

def convertDay(date : str):
    month, day, year = re.findall(r'\w+', date)
    month = month_dict.get(month)
    if day != None and len(day) < 2 : day = f"0{day}"
    if month != None and month/10 < 1: month = f"0{month}"
    return f"{year}-{month}-{day}"

data.Joined = data.Joined.apply(convertDay)
```

```python
data.Joined.value_counts()
```

```python
# Convert Joined to Datatime
data['Joined'] = pd.to_datetime(data['Joined'], format='%Y-%m-%d')
```

### Convert Height from feet-inch to centimeter

```python
data['Height'].value_counts()
```

```python
def convertHeight(height: str):
  if ("\'") not in height: return height
  feet, inch = pd.to_numeric(re.findall(r'\d+', string = height))
  return round(feet*30.48 + inch*2.54)
data['Height'] = data['Height'].apply(convertHeight)
```

```python
data['Height']
```

### Converting Weight from Pound to Kg

```python
data['Weight'].value_counts()
```

```python
def convertWeight(weight: str):
  if 'lbs' not in weight: return weight
  lbs = pd.to_numeric(re.findall(pattern = '\d+', string = weight))[0]
  return round(lbs*0.453592)

data.Weight = data.Weight.apply(convertWeight)
```

```python
def split_team_contract(col=None):
  col = col.str.replace(r'^[\n]{4}','')     # x√≥a 4 k√≠ t·ª± '\n' ·ªü ƒë·∫ßu
  col = col.str.replace(r'[\n]{2}','')      # x√≥a 2 k√≠ t·ª± '\n' ·ªü cu·ªëi
  data[['Team','Contract']] = col.str.split('\n',1,expand=True)     # chia chu·ªói ·ªü v·ªã tr√≠ k√≠ t·ª± '\n'
  return col
data['Team & Contract'] = split_team_contract(data['Team & Contract'])

#drop column Team & Contract
data = data.drop(columns=['Team & Contract'])
```

```python
# Replace '~' to '-'
data['Contract'] = data['Contract'].str.replace('~', '-')
```

### Process Value, Wage and Rease Clause columns

```python
def clean_column(col):
  col = col.lstrip('‚Ç¨')
  end_char = col[-1]  #get the end character
  if end_char == 'M':
    return float(col[:-1]) * 1e6
  elif end_char == 'K':
    return float(col[:-1]) * 1e3
  else:
    return float(col)
```

```python
# excu clean_column function

data['Value'] = data['Value'].apply(clean_column)

data['Wage'] = data['Wage'].apply(clean_column)

data['Release Clause'] = data['Release Clause'].apply(clean_column)
```

### Process W/F, SM and IR columns

```python
# Remove '‚òÖ'
def remove_star(col):
  col = col.rstrip('‚òÖ')
  return int(col)
```

```python
# exc remove_star function
data['W/F'] = data['W/F'].apply(remove_star)

data['SM'] = data['SM'].apply(remove_star)

data['IR'] = data['IR'].apply(remove_star)
```

### Process Hits column


```python
def convert_value(col):
  if isinstance(col, str):
    col = col.lstrip('\n')
    if col.endswith('K'):
      return float(col[:-1]) * 1e3
    else:
      return float(col)
  else:
    return float(col)
```

```python
# exc convert_value function

data['Hits'] = data['Hits'].apply(convert_value)
```
